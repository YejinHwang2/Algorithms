```python
from typing import List
def isValidSudoku(board:List[List[str]])->bool:
    col_set = [set() for i in range(9)]
    sub_set = [set() for i in range(0)]
    for row in range(len(board)):
        
        row_set = set()
        for i in range(len(board[row])):
            num = board[row][i]
            
            #row side check
            if num != "." and num not in row_set:
                row_set.add(num)
            elif num in row_set:
                return False
            
            #col side check
            if num != "." and num not in col_set[i]:
                col_set[i].add(num)
            elif num in col_set[i]:
                return False
            
            #sub-box check
            if num != ".":
                if row in [0,1,2]:
                    if i in [0,1,2]:
                        if num not in sub_set[0]:
                            sub_set[0].add(num)
                        else:
                            return False
                    
                    elif i in [3,4,5]:
                        if num not in sub_set[1]:
                            sub_set[1].add(num)
                        else:
                            return False
                
                    else:
                        if num not in sub_set[2]:
                            sub_set[2].add(num)
                        else:
                            return False
                    
                elif row in [3,4,5]:
                    if i in [0,1,2]:
                        if num not in sub_set[3]:
                            sub_set[3].add(num)
                        else:
                            return False
                    elif i in [3,4,5]:
                        if num not in sub_set[4]:
                            sub_set[4].add(num)
                        else:
                            return False
                    else:
                        if num not in sub_set[5]:
                            sub_set[5].add(num)
                        else:
                            return False
                        
                else:
                    if i in [0,1,2]:
                        if num not in sub_set[6]:
                            sub_set[6].add(num)
                        else:
                            return False
                    elif i in [3,4,5]:
                        if num not in sub_set[7]:
                            sub_set[7].add(num)
                        else:
                            return False
                    else:
                        if num not in sub_set[8]:
                            sub_set[8].add(num)
                        else:
                            return False
        return True
```
